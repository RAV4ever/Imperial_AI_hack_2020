Let's try to do this: taking the sum for all treatments accross all centres for every particular treatment (which corresponds to a different csv file from the dataset) and then create a log with their dates. This can show their trend, when plotted against the dates. 

```
#First of all, see if the csv files are all present (no missing entries) and whether they follow a particular order

import glob
csvall = np.array(glob.glob("drive/My Drive/ai_hack_2020/nhs_data/*.csv"))
```

```
#Now we have all the csv files in the dataset. We want all the numbered ones, but first let's cut the first parts

csvall[0][37:]

len(csvall)
```

```
csvgood = []

for i in range(len(csvall)):

  #print(i)
  temp = csvall[i][37:]
  #print(temp)
  csvgood.append(temp)
  
```

```
#csvgood
```

```
len(csvgood)
```

```
csvgood[0][0]
```

```
csvbest = []

for i in csvgood:

  if i[0] == '0' or i[0] == '1' or i[0] == '2' or i[0] == '3':

    csvbest.append(i)
```

```
len(csvbest)
```

```
csvbest[12]
```

```
allgood = [[] for _ in range(len(csvbest))]
```

```
alldates = [[] for _ in range(len(csvbest))]
```

```
len(alldates)
```

```
#we now have a list with all the good datasets - we want to use this to extract the values and take sums
#for each date across all centres, for each particular csv files (i.e. each treatment)

for i in range(len(csvbest)):

  allgood[i] = pd.read_csv("drive/My Drive/ai_hack_2020/nhs_data/" + str(csvbest[i]))
```

```
allgood[0]
```

```
len(allgood)
```

```
allsum = []

for i in range(len(allgood)):

  temp_sum = []

  for j in range(len(allgood[i])):

    temp = np.asarray(allgood[i].iloc[j,:])
    temp2 = temp[1:]
    tempsum = np.nansum(temp2)
    temp_sum.append(tempsum)

  allsum.append(temp_sum)  

```

```
len(allsum)
```

```
len(alldates)
```

```
numbers = [2]

for i in numbers:

  plt.plot(alldates[i], allsum[i], 'o-', label = str(i))

plt.legend()
#plt.grid()
```
